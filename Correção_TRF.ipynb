```python

# Dicion치rio de mapeamento de n칰meros para nomes dos meses
meses = {
    1: 'janeiro', 2: 'fevereiro', 3: 'mar칞o', 4: 'abril', 5: 'maio', 6: 'junho',
    7: 'julho', 8: 'agosto', 9: 'setembro', 10: 'outubro', 11: 'novembro', 12: 'dezembro'
}

# Substituindo os n칰meros pelos nomes dos meses
data15_24['mes'] = data15_24['mes'].replace(meses)
```

```python
# Substitui칞칫es na coluna 'fase_dia'
data15_24['fase_dia'] = data15_24['fase_dia'].replace({
    'Plena noite': 'Plena Noite',  # corrige a capitaliza칞칚o
    '0': '(null)'                  # substitui '0' por '(null)'
})

print("Valores da coluna 'fase_dia' atualizados com sucesso!")
```

```python
# Faz as substitui칞칫es desejadas
data15_24['condicao_metereologica'] = data15_24['condicao_metereologica'].replace({
    'Nevoeiro/neblina': 'Nevoeiro/Neblina',
    'Ceu Claro': 'C칠u Claro',
    'Ignorada': 'Ignorado',
    '0': '(null)'
})

print("Valores da coluna 'condicao_metereologica' atualizados com sucesso!")
```
