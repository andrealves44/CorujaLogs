```python
arquivo_1 = '/content/drive/My Drive/Data/RENAEST_CSV/RENAEST-2015.csv'
arquivo_2 = '/content/drive/My Drive/Data/RENAEST_CSV/RENAEST-2016.csv'
arquivo_3 = '/content/drive/My Drive/Data/RENAEST_CSV/RENAEST-2017.csv'
arquivo_4 = '/content/drive/My Drive/Data/RENAEST_CSV/RENAEST-2018.csv'
arquivo_5 = '/content/drive/My Drive/Data/RENAEST_CSV/RENAEST-2019.csv'
arquivo_6 = '/content/drive/My Drive/Data/RENAEST_CSV/RENAEST-2020.csv'
arquivo_7 = '/content/drive/My Drive/Data/RENAEST_CSV/RENAEST-2021.csv'
arquivo_8 = '/content/drive/My Drive/Data/RENAEST_CSV/RENAEST-2022.csv'
arquivo_9 = '/content/drive/My Drive/Data/RENAEST_CSV/RENAEST-2023.csv'
arquivo_10 = '/content/drive/My Drive/Data/RENAEST_CSV/RENAEST-2024.csv'
```

```python
rena_2015 = pd.read_csv(arquivo_1, sep = ';', encoding='utf-8', low_memory=False)
rena_2016 = pd.read_csv(arquivo_2, sep = ';', encoding='utf-8', low_memory=False)
rena_2017 = pd.read_csv(arquivo_3, sep = ';', encoding='utf-8', low_memory=False)
rena_2018 = pd.read_csv(arquivo_4, sep = ';', encoding='utf-8', low_memory=False)
rena_2019 = pd.read_csv(arquivo_5, sep = ';', encoding='utf-8', low_memory=False)
rena_2020 = pd.read_csv(arquivo_6, sep = ';', encoding='utf-8', low_memory=False)
rena_2021 = pd.read_csv(arquivo_7, sep = ';', encoding='utf-8', low_memory=False)
rena_2022 = pd.read_csv(arquivo_8, sep = ';', encoding='utf-8', low_memory=False)
rena_2023 = pd.read_csv(arquivo_9, sep = ';', encoding='utf-8', low_memory=False)
rena_2024 = pd.read_csv(arquivo_10, sep = ';', encoding='utf-8', low_memory=False)
```

```python
rena_2020.info()
```

```python
# Lista com todos os dataframes
dataframes = [rena_2015, rena_2016, rena_2017, rena_2018, rena_2019,
              rena_2020, rena_2021, rena_2022, rena_2023, rena_2024]

# Concatenar todos em um √∫nico dataframe
rena_total = pd.concat(dataframes, ignore_index=True, join='outer')
```

```python
rena_total.info()
```

```python
print("üìò Tipos de dados das colunas:")
print(rena_total.dtypes)
```

```python
print("üîé Valores ausentes por coluna:")
print(rena_total.isna().sum())
```

```python
print("üî§ Valores √∫nicos em colunas categ√≥ricas:")
for col in rena_total.select_dtypes(include='object').columns:
    print(f"\nColuna: {col}")
    print(rena_total[col].unique())
```

```python
# Padronizar nomes da coluna 'REGI√ÉO' para letras min√∫sculas
rena_total['REGI√ÉO'] = rena_total['REGI√ÉO'].str.lower()
```

```python
print(rena_total['REGI√ÉO'].unique())
```
