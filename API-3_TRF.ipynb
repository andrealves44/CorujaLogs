```python
data15 = pd.read_csv(arquivo_1, sep=';', low_memory=False, encoding='cp1252')
data16 = pd.read_csv(arquivo_2, sep=';', low_memory=False, encoding='cp1252')
data17 = pd.read_csv(arquivo_3, sep=';', low_memory=False, encoding='cp1252')
data18 = pd.read_csv(arquivo_4, sep=';', low_memory=False, encoding='cp1252')
data19 = pd.read_csv(arquivo_5, sep=';', low_memory=False, encoding='cp1252')
data20 = pd.read_csv(arquivo_6, sep=';', low_memory=False, encoding='cp1252')
data21 = pd.read_csv(arquivo_7, sep=';', low_memory=False, encoding='cp1252')
data22 = pd.read_csv(arquivo_8, sep=';', low_memory=False, encoding='cp1252')
data23 = pd.read_csv(arquivo_9, sep=';', low_memory=False, encoding='cp1252')
data24 = pd.read_csv(arquivo_10, sep=';', low_memory=False, encoding='cp1252')
```

```python
# Remove a coluna 'ano' do DataFrame
data15 = data15.drop('ano', axis=1)
```

```python
# Substituir os nomes dos dias da semana
data15['dia_semana'] = data15['dia_semana'].replace({
    'Segunda': 'segunda-feira',
    'Terça': 'terça-feira',
    'Quarta': 'quarta-feira',
    'Quinta': 'quinta-feira',
    'Sexta': 'sexta-feira',
    'Sábado': 'sábado',
    'Domingo': 'domingo'
})
```

```python
# Converter a coluna 'data_inversa' para o formato correto
data16['data_inversa'] = pd.to_datetime(
    data16['data_inversa'],
    format='%d/%m/%y'  # formato original
).dt.strftime('%Y-%m-%d').astype('object')  # formato desejado
```

```python
# Substituir os nomes dos dias da semana
data16['dia_semana'] = data16['dia_semana'].replace({
    'Segunda': 'segunda-feira',
    'Terça': 'terça-feira',
    'Quarta': 'quarta-feira',
    'Quinta': 'quinta-feira',
    'Sexta': 'sexta-feira',
    'Sábado': 'sábado',
    'Domingo': 'domingo'
})
```

```python
data16.head()
```

```python
data16.info()
```

```python
# Juntar todos os dataframes de data15 até data24
data_total = pd.concat(
    [data15, data16, data17, data18, data19, data20, data21, data22, data23, data24],
    ignore_index=True,  # reorganiza os índices
    join='outer'        # mantém todas as colunas (mesmo que faltem em alguns)
)

data_total.to_csv('data15-24.csv', index=False, encoding='utf-8', sep=';')
```

```python
print("Total de valores NaN antes da substituição:")
print(data_total.isna().sum())
```

```python
# Substituir todos os valores NaN por 0
data_total = data_total.fillna(0)
```

```python
print("Total de valores NaN depois da substituição:")
print(data_total.isna().sum())
```

```python
# Criar uma nova coluna 'ano' a partir de 'data_inversa'
data_total['ano'] = pd.to_datetime(data_total['data_inversa']).dt.year
```

```python
# Criar uma nova coluna 'mes' a partir de 'data_inversa'
data_total['mes'] = pd.to_datetime(data_total['data_inversa']).dt.month
```
